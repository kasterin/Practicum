```python
import re 
import pandas as pd 

with open('Уэллс_-_Война_миров.txt', 'r') as f: 
    data = f.read() 
prepositions = ['и', 'в', 'на', 'не', 'но', 'с', 'к', 'до', 'у', 'о', 'за', 'из', 'по', 'от', 'а', 'же', 'то', 'В']
words = [word for word in re.findall(r'\b[\w-]+\b', data, flags=re.IGNORECASE) if word.lower() not in prepositions] 
word_counts = pd.Series(words).value_counts() 
top_words = word_counts.head(20) 
print(top_words) 
top_words.plot(kind='bar') 
num_chars = len(data) 
num_chars_nospace = len(data.replace(' ', '')) 
num_words = len(words) 
num_lines = data.count('\n') + 1 
print(f"Строки: {num_lines}")
print(f"Слова: {num_words}") 
print(f"Символы с пробелами: {num_chars}") 
print(f"Символы без пробелов: {num_chars_nospace}") 
```
